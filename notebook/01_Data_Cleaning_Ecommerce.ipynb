```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../dataset/raw/data.csv", encoding='latin1')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=["CustomerID"])
```

```python
df.isnull().sum()
```

```python
df= df[~df["InvoiceNo"].astype(str).str.startswith('C')]
```

```python
df= df[df["Quantity"]>0]
df= df[df["UnitPrice"]>0]
```

```python
df["InvoiceDate"]=pd.to_datetime(df["InvoiceDate"])
```

```python
df.dtypes
```

```python
df["Revenue"]=df["Quantity"]*df["UnitPrice"]
```

```python
df=df.drop_duplicates()
```

```python
df.shape
```

```python
df.to_csv("../dataset/processed/clean_ecommerce.csv", index=False)
```
