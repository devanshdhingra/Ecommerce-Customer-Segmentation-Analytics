```python
import pandas as pd
```

```python
df= pd.read_csv("../dataset/processed/clean_ecommerce.csv")
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
df.head()
```

```python
customer_df= df.groupby("CustomerID").agg({
    "InvoiceNo":"nunique",
    "Revenue":"sum",
    "Quantity":"sum",
    "InvoiceDate":["min","max"]
}).reset_index()
```

```python
customer_df.columns = [
    "CustomerID",
    "Total_Orders",
    "Total_Revenue",
    "Total_Quantity",
    "First_Purchase_Date",
    "Last_Purchase_Date"
]
```

```python
customer_df["Avg_Order_Value"] = (
    customer_df["Total_Revenue"] / customer_df["Total_Orders"]
)
```

```python
customer_df.head()
customer_df.shape
```

```python
customer_df.to_csv("../dataset/processed/customer_features.csv", index=False)
```
