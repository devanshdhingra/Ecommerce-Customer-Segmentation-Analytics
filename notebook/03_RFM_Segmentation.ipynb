```python
import pandas as pd

customer_df = pd.read_csv("../dataset/processed/customer_features.csv")
```

```python
customer_df["Last_Purchase_Date"] = pd.to_datetime(customer_df["Last_Purchase_Date"])
```

```python
today = customer_df["Last_Purchase_Date"].max()+pd.Timedelta(days=1)
```

```python
customer_df["Recency"]=(today-customer_df["Last_Purchase_Date"]).dt.days
```

```python
customer_df["Frequency"] = customer_df["Total_Orders"]
```

```python
customer_df["Monetary"] = customer_df["Total_Revenue"]
```

```python
customer_df[["Recency","Frequency","Monetary"]].head()
```

```python
customer_df["R_score"] = pd.qcut(customer_df["Recency"], 4, labels=[4,3,2,1])
```

```python
customer_df["F_score"] = pd.qcut(
    customer_df["Frequency"],
    4,
    duplicates="drop"
)

customer_df["F_score"] = customer_df["F_score"].cat.codes + 1
```

```python
customer_df["M_score"] = pd.qcut(customer_df["Monetary"], 4, labels=[1,2,3,4])
```

```python
customer_df["RFM_Score"] = (
    customer_df["R_score"].astype(int) +
    customer_df["F_score"].astype(int) +
    customer_df["M_score"].astype(int)
)
```

```python
def segment(score):
    if score >= 10:
        return "High Value"
    elif score >= 6:
        return "Medium Value"
    else:
        return "Low Value"

customer_df["Segment"] = customer_df["RFM_Score"].apply(segment)
```

```python
customer_df.to_csv("../dataset/processed/rfm_customer_segments.csv", index=False)
```
